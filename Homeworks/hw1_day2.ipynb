```
singleDigit=int(input("Enter a singel digit number: "))
if (singleDigit > 0 and singleDigit < 10):
  nums=list(range(singleDigit+1))
  print (nums)
else:
  print ("It is not a single digirt")

half= int(len(nums)/2)
firstHalf=list(range(half))
secondHalf=nums[half:singleDigit+1]
secondList=secondHalf + firstHalf
print (secondList)
```
