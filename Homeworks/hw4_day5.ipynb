```
class Animals:
  def __init__(self, name, type, color, age):
    self.names=name
    self.types=type
    self.colors=color
    self.ages=age

class Cats(Animals):
  animal="cat"
  def print_info(self):
    print (self.names + " is a "  + self.animal +  ". It's " + self.ages + " year old a/an " + self.colors + " " +self.types + "\n")


class Dogs(Animals):
  animal="dog"
  def print_info(self):
    print (self.names + " is a "  + self.animal +  ". It's " + self.ages + " year old a/an " + self.colors + " " +self.types + "\n")


cat1=Cats("duman","scootish fold","gray","2")
cat1.print_info()


dog1=Dogs("Max", "golden", "brown", "4")
dog1.print_info()
```
