```
#Q1

correctid="username"
correctpass="password"
username=input("Enter your username:")
password=input("Enter your password")

if username==correctid and password==correctpass:
  print ("You are logged in")
elif username != correctid:
  print ("There are not any user named " +username)
elif username==correctid and password!=correctpass:
  print ("Wrong password")
```
