```

def findPrimes(num):
  for z in range(1,(num+1)):
    if z==2:
     primes=[2]
    for i in range (2,z):
      if(z%i)==0:
        break             
      else:
        if z in primes:
          break
        else:
          primes.append(z)
  print (primes)
        
a=int(input("Enter a number: "))
findPrimes (a)

```
